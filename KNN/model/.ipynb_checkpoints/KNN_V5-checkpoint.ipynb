```python
# -*- coding: UTF-8 -*-

from __future__ import division      #除数可以显示为float

from six import StringIO    #使用聚宽readfile函数
import numpy as np
import pandas as pd

import time                 #使用time stamp
import datetime             

import matplotlib.pyplot as plt

import math

# 最基本的KNN算法
from sklearn.neighbors import KNeighborsClassifier

# 导入样本拆分模块
from sklearn.model_selection import train_test_split

# 导入KNN半径算法
from sklearn.neighbors import RadiusNeighborsClassifier

# 评分函数
from sklearn.metrics import precision_score, recall_score, f1_score

# 交叉评分函数
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
Data_Csv = pd.DataFrame()

for file_i in range(1,11,1):
    
    file_name = "Data_Washed%d.csv"%(int(file_i))
    
    Data_Csv = pd.concat([ Data_Csv , pd.read_csv(file_name) ] )


Col_Name = ['EMA_5','EMA_5_pre1','EMA_5_pre2','EMA_10','EMA_10_pre1','EMA_10_pre2','EMA_gap','EMA_gap_pre1','EMA_gap_pre2','KDJ_K','KDJ_K_pre1','KDJ_K_pre2','KDJ_D','KDJ_D_pre1','KDJ_D_pre2','KDJ_J','KDJ_J_pre1','KDJ_J_pre2','RSI','RSI_pre1','RSI_pre2','MACD_dif','MACD_dif_pre1','MACD_dif_pre2','MACD_dea','MACD_dea_pre1','MACD_dea_pre2','MACD_macd','MACD_macd_pre1','MACD_macd_pre2','MOM_12','MOM_12_pre1','MOM_12_pre2','MOM_25','MOM_25_pre1','MOM_25_pre2','MOM_gap','MOM_gap_pre1','MOM_gap_pre2','Long_Short_Rate_OBV','Long_Short_Rate_OBV_pre1','Long_Short_Rate_OBV_pre2','Volume','Volume_pre1','Volume_pre2']
```

```python
Data_Csv = Data_Csv.fillna(0)

print Data_Csv.shape
Stock_list = set(list(Data_Csv['code']))
```

```python
# 根据上面的结果
# 设置k的系数
k = 11
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline

# 增加多项式预处理
def polynomial_model(degree=1, **kwarg):
    polynomial_features = PolynomialFeatures(degree=degree,
                                             include_bias=False)
    knn_regression = KNeighborsClassifier(n_neighbors=k)
    pipeline = Pipeline([("polynomial_features", polynomial_features),
                         ("knn_regression", knn_regression)])
    return pipeline
```

```python
score_stock_pd = pd.DataFrame()
score_stock_dict = {}
score_stock_dict_regression = {}

attempt_count = 0
success_count = 0
for stock_name in Stock_list:
    # 测试数据
    X =  np.array(Data_Csv.loc[Data_Csv['code'] == stock_name,Col_Name])
    
    # 测试结果
    y = np.array(Data_Csv.loc[Data_Csv['code'] == stock_name,"win_rate"])

    # 分别生成训练数据、测试数据
    x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
    
    # 单一模型预测
    # 训练模型
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    
    # 多项式模型预测
    model_regression = polynomial_model(degree=2, penalty='l1')
    model_regression.fit(x_train, y_train)
    y_pred_regression = model_regression.predict(x_test);
    
    
#     print "判断正例次数：%d"%int(len(y_test[y_pred_total == 1]))
#     print "实际正例次数：%d"%int(sum(y_test[y_pred_total == 1] ==1))
#     print "正确率为：%f"%float(float(sum(y_test[y_pred_total == 1] ==1))/float(len(y_test[y_pred_total == 1])))


    attempt_count += int(len(y_test[y_pred == 1]))
    success_count += int(sum(y_test[y_pred == 1] ==1))
    
    score_stock_dict[stock_name] = [precision_score(y_test,y_pred ,average='binary'),int(len(y_test[y_pred == 1])),int(sum(y_test[y_pred == 1] ==1))]
    
    score_stock_dict_regression[stock_name] = [precision_score(y_test,y_pred_regression ,average='binary'),int(len(y_test[y_pred_regression == 1])),int(sum(y_test[y_pred_regression == 1] ==1))]

    print "正常回归：股票名称: %s; 正确率:%f；正例个数:%d；正确个数:%d"%(str(stock_name),score_stock_dict[stock_name][0],score_stock_dict[stock_name][1],score_stock_dict[stock_name][2])
    print "多项式回归：股票名称: %s; 正确率:%f；正例个数:%d；正确个数:%d"%(str(stock_name),score_stock_dict_regression[stock_name][0],score_stock_dict_regression[stock_name][1],score_stock_dict_regression[stock_name][2])

print attempt_count
print success_count
print "尝试总次数：%d；正确次数：%d，正确率：%f"%(attempt_count,success_count,float(success_count/attempt_count))
```

```python
score_stock_pd = pd.DataFrame.from_dict(score_stock_dict, orient='index',columns = ['TP','Predict_counts','Correct_counts'])

score_stock_pd.to_csv("stock_rate.csv")

score_stock_pd = pd.DataFrame.from_dict(score_stock_dict_regression, orient='index',columns = ['TP','Predict_counts','Correct_counts'])

score_stock_pd.to_csv("stock_rate_regression.csv")
```

```python

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

    
cross_score_dict = {}

kfold = KFold(n_splits=10)

cv_result = cross_val_score(KNeighborsClassifier(n_neighbors=k), X, y, cv=kfold)

cross_score_dict["Total"] = float(cv_result.mean())

print "总体数据交叉评分为：%f"% (float(cv_result.mean()))


for stock_name in Stock_list:
    # 测试数据
    X =  np.array(Data_Csv.loc[Data_Csv['code'] == stock_name,Col_Name])
    
    # 测试结果
    y = np.array(Data_Csv.loc[Data_Csv['code'] == stock_name,"win_rate"])

    cv_result = cross_val_score(KNeighborsClassifier(n_neighbors=k), X, y, cv=kfold)

    cross_score_dict[str(stock_name)] = float(cv_result.mean())

    print "股票名称: %s; 交叉评分为:%f"%(str(stock_name),float(cv_result.mean()))
```
