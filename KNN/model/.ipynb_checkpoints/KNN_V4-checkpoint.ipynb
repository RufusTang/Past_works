```python
from __future__ import division      #除数可以显示为float

from six import StringIO    #使用聚宽readfile函数
import numpy as np
import pandas as pd

import time                 #使用time stamp
import datetime             

import matplotlib.pyplot as plt

import math

# 最基本的KNN算法
from sklearn.neighbors import KNeighborsClassifier

# 导入样本拆分模块
from sklearn.model_selection import train_test_split

# 导入KNN半径算法
from sklearn.neighbors import RadiusNeighborsClassifier

# 评分函数
from sklearn.metrics import precision_score, recall_score, f1_score

# 交叉评分函数
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
Data_Csv = pd.DataFrame()

for file_i in range(1,11,1):
    
    file_name = "Data_Washed%d.csv"%(int(file_i))
    
    Data_Csv = pd.concat([ Data_Csv , pd.read_csv(file_name) ] )


Col_Name = ['EMA_5','EMA_5_pre1','EMA_5_pre2','EMA_10','EMA_10_pre1','EMA_10_pre2','EMA_gap','EMA_gap_pre1','EMA_gap_pre2','KDJ_K','KDJ_K_pre1','KDJ_K_pre2','KDJ_D','KDJ_D_pre1','KDJ_D_pre2','KDJ_J','KDJ_J_pre1','KDJ_J_pre2','RSI','RSI_pre1','RSI_pre2','MACD_dif','MACD_dif_pre1','MACD_dif_pre2','MACD_dea','MACD_dea_pre1','MACD_dea_pre2','MACD_macd','MACD_macd_pre1','MACD_macd_pre2','MOM_12','MOM_12_pre1','MOM_12_pre2','MOM_25','MOM_25_pre1','MOM_25_pre2','MOM_gap','MOM_gap_pre1','MOM_gap_pre2','Long_Short_Rate_OBV','Long_Short_Rate_OBV_pre1','Long_Short_Rate_OBV_pre2','Volume','Volume_pre1','Volume_pre2']
```

```python
Data_Csv.loc[Data_Csv["win_rate"] == -1].shape[0]
Data_Csv.loc[Data_Csv["win_rate"] == 1].shape[0]
Data_Csv.loc[Data_Csv["win_rate"] == 0].shape[0]

```

```python
Data_Csv = Data_Csv.fillna(0)

print Data_Csv.shape
Stock_list = set(list(Data_Csv['code']))
```

分别看各个股票的预测情况

```python
# 根据上面的结果
# 设置k的系数
k = 11
```

```python
# 测试数据
X =  np.array(Data_Csv.loc[:,Col_Name])
# 测试结果
y = np.array(Data_Csv.loc[:,"win_rate"])

# 分别生成训练数据、测试数据
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)

model_total =  KNeighborsClassifier(n_neighbors=k)

model_total.fit(x_train, y_train)

y_pred_total = model_total.predict(x_test)

print "k为：%d，整体正例评分:%f"%(int(k),precision_score(y_test,y_pred_total ,average='binary'))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline

# 增加多项式预处理
def polynomial_model(degree=1, **kwarg):
    polynomial_features = PolynomialFeatures(degree=degree,
                                             include_bias=False)
    knn_regression = KNeighborsClassifier(n_neighbors=k)
    pipeline = Pipeline([("polynomial_features", polynomial_features),
                         ("knn_regression", knn_regression)])
    return pipeline

model_regression = polynomial_model(degree=2, penalty='l1')

model_regression.fit(x_train, y_train)

y_pred_regression = model_regression.predict(x_test);
    
print precision_score(y_test,y_pred_regression ,average='binary')
```

```python
score_stock_pd = pd.DataFrame()
score_stock_dict = {}

for stock_name in Stock_list:
    # 测试数据
    X =  np.array(Data_Csv.loc[Data_Csv['code'] == stock_name,Col_Name])
    
    # 测试结果
    y = np.array(Data_Csv.loc[Data_Csv['code'] == stock_name,"win_rate"])

    # 分别生成训练数据、测试数据
    x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
    
    # 训练模型
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(x_train, y_train)

    model_regression = polynomial_model(degree=2, penalty='l1')
    model_regression.fit(x_train, y_train)
    
    
    # 单一模型预测
    y_pred = model.predict(x_test)
    
    # 多项式模型预测
    y_pred__regression = model_regression.predict(x_test);
    
    # 共性模型预测
    y_pred_total = model_total.predict(x_test)
    
     
    score_stock_dict[stock_name] = [precision_score(y_test,y_pred_total ,average='binary'),precision_score(y_test,y_pred__regression ,average='binary'),precision_score(y_test,y_pred ,average='binary'),sum(np.array(y_pred_total) == 1),sum(np.array(y_pred__regression) == 1),sum(np.array(y_pred) == 1)]
    
    print "股票名称: %s; 共性模型评分:%f；多项式个性模型评分：%f；个性模型评分:%f"%(str(stock_name),score_stock_dict[stock_name][0],score_stock_dict[stock_name][1],score_stock_dict[stock_name][2])
```

```python
score_stock_pd = pd.DataFrame.from_dict(score_stock_dict, orient='index')

score_stock_pd.to_csv("stock_rate.csv")
```

```python

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

    
cross_score_dict = {}

kfold = KFold(n_splits=10)

cv_result = cross_val_score(KNeighborsClassifier(n_neighbors=k), X, y, cv=kfold)

cross_score_dict["Total"] = float(cv_result.mean())

print "总体数据交叉评分为：%f"% (float(cv_result.mean()))


for stock_name in Stock_list:
    # 测试数据
    X =  np.array(Data_Csv.loc[Data_Csv['code'] == stock_name,Col_Name])
    
    # 测试结果
    y = np.array(Data_Csv.loc[Data_Csv['code'] == stock_name,"win_rate"])

    cv_result = cross_val_score(KNeighborsClassifier(n_neighbors=k), X, y, cv=kfold)

    cross_score_dict[str(stock_name)] = float(cv_result.mean())

    print "股票名称: %s; 交叉评分为:%f"%(str(stock_name),float(cv_result.mean()))
```

```python
cross_score_stock_pd = pd.DataFrame.from_dict(cross_score_dict, orient='index')

cross_score_stock_pd.to_csv("cross_score_rate.csv")
```
